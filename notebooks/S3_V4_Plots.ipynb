# Matplotlib Plot 

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x,y)
plt.title("F(x,y)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# theta from 0 to 360 degree.

theta = np.linspace(0,2*np.pi,30) 
y = 10*np.sin(theta)

plt.plot((180/np.pi)*theta,y,'r--o')

plt.title("Sine Wave")
plt.xlabel("theta")
plt.ylabel("Y")
plt.show()
```

# Multiple Plots

```python
import numpy as np
import matplotlib.pyplot as plt

# theta from 0 to 360 degree.

theta = np.linspace(0,2*np.pi,30) 

y1 = 10*np.sin(theta)
y2 = 5* np.sin(theta)

plt.plot((180/np.pi)*theta,y1,'r--o',label="y1")
plt.plot((180/np.pi)*theta,y2,'g-*',label="y2")
plt.legend()

plt.xlim(0,360)
plt.ylim(-10,10)
plt.grid()
plt.title("Sine Wave",fontsize=15)
plt.xlabel("theta")
plt.ylabel("Y")
plt.show()
```

# Subplots

```python
import numpy as np
import matplotlib.pyplot as plt

# theta from 0 to 360 degree.

theta = np.linspace(0,2*np.pi,30) 

y1 = 10*np.sin(theta)
y2 = 5* np.sin(theta)

plt.subplot(2,1,1)
plt.plot((180/np.pi)*theta,y1,'r--o',label="y1")
plt.xlim(0,360)
plt.ylim(-10,10)
plt.grid()
plt.title("Sine Wave",fontsize=15)
plt.xlabel("theta")
plt.ylabel("Y")



plt.subplot(2,1,2)
plt.plot((180/np.pi)*theta,y2,'g-*',label="y2")
plt.legend()
plt.xlim(0,360)
plt.ylim(-5,5)
plt.grid()
plt.title("Sine Wave",fontsize=15)
plt.xlabel("theta")
plt.ylabel("Y")
plt.tight_layout(pad = 2.0)
plt.show()
```

# Surface Plot

```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
ax = Axes3D(fig)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet)
fig.colorbar(surf)

plt.show()
```
