## Dynamic Magnification Factor

\begin{equation}
DMF = \frac{1}{\sqrt{(1-\beta^2)^2 + (2\xi\beta)^2}}
\end{equation}

<ul>
<li>Let's assume a frequency ratio between 0 and 2.</li>    
<li>Initially assume 5% damping ratio, $\xi = 0.05$</li>    
</ul>

```python
# Dependencies
import math #Basic math library
import numpy as np #Numpy -> tools for working with arrays
import matplotlib.pyplot as plt #Matplotlib -> tools for plotting

#Generate a vector of frequency ratios
max_B = 2 #Max value of Beta
delta_B = 0.01 #Step size in range of Beta values
nPoints = max_B/delta_B #Number of datapoints in the range
B = np.linspace(0,max_B, int(nPoints)) #Vector of Beta values
xi = 0.05 #Damping ratio

DMF = 1/np.sqrt((1-B**2)**2 + (2*xi*B)**2) #Dynamic Magnification Factor

#Plotting
fig = plt.figure() 
axes = fig.add_axes([0.1,0.1,2,1.5]) 
axes.plot(B,DMF, label = "5% damping")
axes.set_xlim([0,max_B])
axes.set_xlabel('Frequency ratio $\\beta$')
axes.set_ylabel('Dynamic Magnification Factor')
axes.set_title('Dynamic Magnification Factor')
axes.grid()
axes.legend()
plt.show()
```

### Observations

<ul>
    <li>When $\omega$ is close to $\omega_n$ the DMF peaks
        <li> DMF peaks at approximately 10 for 5% damping - impact of dynamic over static force
</ul>   

<b>What about different damping ratios??</b>

### Consider a range of damping ratios

```python
Xi = [0.05, 0.1, 0.2, 0.5] #List of damping ratios

#Set up the figure 
fig = plt.figure() 
axes = fig.add_axes([0.1,0.1,2,1.5]) 
axes.set_xlim([0,max_B])
axes.set_xlabel('Frequency ratio $\\beta$')
axes.set_ylabel('Dynamic Magnification Factor')
axes.set_title('Dynamic Magnification Factor')
axes.grid()

#For each damping ratio in the list, add a plot to the figure
for xi in Xi:
    damping = xi*100 #percentage damping (for label)
    
    DMF = 1/np.sqrt((1-B**2)**2 + (2*xi*B)**2)
    
    axes.plot(B,DMF, label = "{A}% damping".format(A=damping))

    axes.legend()

plt.savefig('DMF.png', bbox_inches='tight')
```

### Observations

<ul>
   <li> Dynamic impact is reduced as the level of structural damping increases but remember many civil structures have low ($<5%$) levels of damping - $\therefore$ dynamic loads can be a big problem
       <li> Peak DMF does not occur at $\beta= 1$, but slightly below - for low levels of damping this offset is often ignored.
</ul>
